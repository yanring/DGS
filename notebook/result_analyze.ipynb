```python
import numpy as np
import pandas as pd
import pylab as pl
from pylab import *
import plotly.offline as py
#import plotly.offline as py
import plotly.graph_objs as go
import plotly.figure_factory as ff
py.init_notebook_mode(connected=True)
```

```python
def prepare_data(path):
    df = pd.read_csv(path)
    df['segament'] = (df['index']/30).astype('int')
    df.timestamp=pd.to_datetime(df.timestamp)
    df['time_cost']=df.timestamp-df.iloc[0].timestamp
    df.time_cost=df.time_cost/ pd.offsets.Minute(1)
    return df.copy()
single_node = prepare_data('/share/distbelief/log/single.csv')
ds_node_1 = prepare_data('/share/distbelief/log/node1_v1_2woker.csv')
ds_node_2 = prepare_data('/share/distbelief/log/node2_v1_2woker.csv')
```

# V1 2woker

```python
# 每100个点平均一次，不然数据太多没法画图
single_node_train_data = single_node.groupby(['segament']).mean()
ds_node_1_train_data = ds_node_1.groupby(['segament']).mean()
ds_node_2_train_data = ds_node_2.groupby(['segament']).mean()
```

```python
# 查看train accuracy

singlenode = go.Scatter(
    x=single_node_train_data.index,
    y=single_node_train_data.training_accuracy,
    name='singlenode'
)
dsnode_1 = go.Scatter(
    x=ds_node_1_train_data.index,
    y=ds_node_1_train_data.training_accuracy,
    name='dsnode_1',
)
dsnode_2 = go.Scatter(
    x=ds_node_2_train_data.index,
    y=ds_node_2_train_data.training_accuracy,
    name='dsnode_2',
)
data = [singlenode,dsnode_1,dsnode_2]
py.iplot(data)
```

```python
# 查看train loss

singlenode = go.Scatter(
    x=single_node_train_data.index,
    y=single_node_train_data.training_loss,
    name='singlenode'
)
dsnode_1 = go.Scatter(
    x=ds_node_1_train_data.index,
    y=ds_node_1_train_data.training_loss,
    name='dsnode_1',
)
dsnode_2 = go.Scatter(
    x=ds_node_2_train_data.index,
    y=ds_node_2_train_data.training_loss,
    name='dsnode_2',
)
data = [singlenode,dsnode_1,dsnode_2]
py.iplot(data)
```
